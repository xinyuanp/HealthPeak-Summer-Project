## Change CBSA codes for 11 primary markets

#### 11 primary markets are seperated into smaller divisions in StratoDem. Sum those divisions to get the original cbsa.

```python
#df_cbsa = pd.read_excel('cbsa.xlsx')
#df_cbsa[(df_cbsa.stra!=0)]

#NIC_CBSA 14460 = [14454, 15764, 40484]
#NIC_CBSA 16980 = [16974, 20994, 23844, 29404]
#NIC_CBSA 19100 = [19124, 23104]
#NIC_CBSA 19820 = [19804, 47664]
#NIC_CBSA 31080 = [11244, 31084]
#NIC_CBSA 33100 = [22744, 33124, 48424]
#NIC_CBSA 35620 = [20524, 35004, 35084, 35614]
#NIC_CBSA 37980 = [15804, 33874, 37964, 48864]
#NIC_CBSA 41860 = [36084, 41884, 42034]
#NIC_CBSA 42660 = [42644, 45104]
#NIC_CBSA 47900 = [43524, 47894]

a1 = [14454, 15764, 40484]
a2 = [16974, 20994, 23844, 29404]
a3 = [19124, 23104]
a4 = [19804, 47664]
a5 = [11244, 31084]
a6 = [22744, 33124, 48424]
a7 = [20524, 35004, 35084, 35614]
a8 = [15804, 33874, 37964, 48864]
a9 = [36084, 41884, 42034]
a10 = [42644, 45104]
a11 = [43524, 47894]
a = a1+a2+a3+a4+a5+a6+a7+a8+a9+a10+a11

#Define the function that group by year only
def merge_code(df):
    
    y1 = df[df.CBSA.isin(a1)].groupby(['YEAR']).sum().reset_index()
    y2 = df[df.CBSA.isin(a2)].groupby(['YEAR'], as_index=False).sum()
    y3 = df[df.CBSA.isin(a3)].groupby(['YEAR'], as_index=False).sum()
    y4 = df[df.CBSA.isin(a4)].groupby(['YEAR'], as_index=False).sum()
    y5 = df[df.CBSA.isin(a5)].groupby(['YEAR'], as_index=False).sum()
    y6 = df[df.CBSA.isin(a6)].groupby(['YEAR'], as_index=False).sum()
    y7 = df[df.CBSA.isin(a7)].groupby(['YEAR'], as_index=False).sum()
    y8 = df[df.CBSA.isin(a8)].groupby(['YEAR'], as_index=False).sum()
    y9 = df[df.CBSA.isin(a9)].groupby(['YEAR'], as_index=False).sum()
    y10 = df[df.CBSA.isin(a10)].groupby(['YEAR'], as_index=False).sum()
    y11 = df[df.CBSA.isin(a11)].groupby(['YEAR'], as_index=False).sum()
    
    y1 = y1.assign(CBSA='14460')
    y2 = y2.assign(CBSA='16980')
    y3 = y3.assign(CBSA='19100')
    y4 = y4.assign(CBSA='19820')
    y5 = y5.assign(CBSA='31080')
    y6 = y6.assign(CBSA='33100')
    y7 = y7.assign(CBSA='35620')
    y8 = y8.assign(CBSA='37980')
    y9 = y9.assign(CBSA='41860')
    y10 = y10.assign(CBSA='42660')
    y11 = y11.assign(CBSA='47900')
    df_delete = df[~df.CBSA.isin(a)]
    
    df_merged_cbsa = df_delete.append([y1,y2,y3,y4,y5,y6,y7,y8,y9,y10,y11], ignore_index=True)
    df_merged_cbsa['CBSA'] = df_merged_cbsa['CBSA'].astype(int)

    return df_merged_cbsa
```

```python
#Define the function that group by year and ages at the same time
def merge_code_age(df):
    
    y1 = df[df.CBSA.isin(a1)].groupby(['YEAR','AGE_G']).sum().reset_index()
    y2 = df[df.CBSA.isin(a2)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y3 = df[df.CBSA.isin(a3)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y4 = df[df.CBSA.isin(a4)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y5 = df[df.CBSA.isin(a5)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y6 = df[df.CBSA.isin(a6)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y7 = df[df.CBSA.isin(a7)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y8 = df[df.CBSA.isin(a8)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y9 = df[df.CBSA.isin(a9)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y10 = df[df.CBSA.isin(a10)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    y11 = df[df.CBSA.isin(a11)].groupby(['YEAR','AGE_G'], as_index=False).sum()
    
    y1 = y1.assign(CBSA='14460')
    y2 = y2.assign(CBSA='16980')
    y3 = y3.assign(CBSA='19100')
    y4 = y4.assign(CBSA='19820')
    y5 = y5.assign(CBSA='31080')
    y6 = y6.assign(CBSA='33100')
    y7 = y7.assign(CBSA='35620')
    y8 = y8.assign(CBSA='37980')
    y9 = y9.assign(CBSA='41860')
    y10 = y10.assign(CBSA='42660')
    y11 = y11.assign(CBSA='47900')
    df_delete = df[~df.CBSA.isin(a)]
    
    df_merged_cbsa = df_delete.append([y1,y2,y3,y4,y5,y6,y7,y8,y9,y10,y11], ignore_index=True)
    df_merged_cbsa['CBSA'] = df_merged_cbsa['CBSA'].astype(int)

    return df_merged_cbsa
```

```python
#Define the function that no need of groupby
def merge_code_nogroup(df):
    
    y1 = df[df.CBSA.isin(a1)].sum()
    y2 = df[df.CBSA.isin(a2)].sum()
    y3 = df[df.CBSA.isin(a3)].sum()
    y4 = df[df.CBSA.isin(a4)].sum()
    y5 = df[df.CBSA.isin(a5)].sum()
    y6 = df[df.CBSA.isin(a6)].sum()
    y7 = df[df.CBSA.isin(a7)].sum()
    y8 = df[df.CBSA.isin(a8)].sum()
    y9 = df[df.CBSA.isin(a9)].sum()
    y10 = df[df.CBSA.isin(a10)].sum()
    y11 = df[df.CBSA.isin(a11)].sum()
    
    df_add =pd.DataFrame({"CBSA":['14460', '16980','19100','19820','31080','33100','35620','37980','41860','42660','47900'], 
                         "AREA":[y1[1], y2[1], y3[1], y4[1], y5[1], y6[1], y7[1], y8[1], y9[1], y10[1], y11[1]]}) 

    df_delete = df[~df.CBSA.isin(a)]
    df0 = df_delete.append(df_add)
    df0['CBSA'] = df0['CBSA'].astype(int)

    return df0
```

```python
#Population Density

#Calculating the population
df_pop_raw = SDAPIQuery.query_api_df(
    query_params=APIQueryParams(
        query_type='COUNT',
        table='populationforecast_metro_annual_population',
        data_fields=('year', 'cbsa', {'population': 'population'}),
        data_filters=(
            LessThanFilter(var='year', val=2030),
            GreaterThanOrEqualToFilter(var='year', val=2007),
        ),
        aggregations=(dict(aggregation_func='sum', variable_name='population'),),
        groupby=('cbsa', 'year'),
        order=('year',),
    )
)
#Matching NIC and StratoDem CBSA
df_pop = merge_code(df_pop_raw)

#Calculating the areas
df_area_raw = SDAPIQuery.query_api_df(
    query_params=APIQueryParams(
            query_type='AREA',
            table='geocookbook_metro_na_shapes_full',
            data_fields=('cbsa', 'area', 'name'),
            data_filters=(),
            groupby=('cbsa', 'name'),
            aggregations=(),
    )
)

#Matching NIC and StratoDem CBSA
df_area = merge_code_nogroup(df_area_raw.drop(columns=['NAME']))
df = df_pop.merge(df_area, on= 'CBSA')
df['population_density'] = df['POPULATION']/df['AREA']
df1 = df.iloc[:,0:2]
df2 = df.iloc[:,4:]
df_den = df1.join(df2)
df_den
#df_den.to_csv('population_density.csv', index=False)
```
